```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# u(t): Unit Step Signal
def u(t: float) -> int:
    return 1 if t >= 0 else 0

# x(t): Signal given for part (a)
def x(t: float) -> float:
    return np.cos(t) * u(t)

# h(t): Impulse response of x1(t)
def h(t: float) -> float:
    return 1/4 * (np.exp(-2*t) - np.exp(-t)) * u(t)
```

```python
# Plot x(t), h(t), and h(-t)

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)

T = np.arange(-20.0, 20.01, 0.01)

axes[0].step(T, [x(t) for t in T])
axes[1].step(T, [h(t) for t in T])
axes[2].step(T, [h(-t) for t in T])

axes[0].set_title("Question: 1 Part (a)")
axes[0].set_ylabel("x(t)")
axes[1].set_ylabel("h(t)")
axes[2].set_ylabel("h(-t)")
axes[2].set_xlabel("t")
```

```python
# Plotting x(t), h(t), and y(t) = x(t) * h(t) for t = [0, 20]

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].set_xticks(range(21))
    axes[i].grid(True)

dt = 0.01
T = np.arange(0, 20.01, dt)
X = [x(t) for t in T]
H = [h(t) for t in T]
ravelH = np.ravel(H)

Y = np.zeros((len(T), 1))
for t in T:
    x2 = np.cos(t-T)
    x2[t-T <= 0] = 0 
    x2 = np.ravel(x2)
    Y[t == T] = np.sum(x2*ravelH) * dt

axes[0].plot(T, X)
axes[1].plot(T, H)
axes[2].plot(T, Y)

axes[0].set_title("Question: 1 Part (a)")
axes[0].set_ylabel("x(t)")
axes[1].set_ylabel("h(t)")
axes[2].set_ylabel("y(t)")
axes[2].set_xlabel("t")
```
