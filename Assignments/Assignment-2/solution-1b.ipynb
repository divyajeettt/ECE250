```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# u(t): Unit Step Signal
def u(t: float) -> int:
    return 1 if t >= 0 else 0

# x(t): Signal given for part (b)
def x(t: float) -> float:
    return np.exp(-t) * np.sin(t) * u(t)

# h(t): Impulse response of x2(t)
def h(t: float) -> float:
    return 1/2 * (np.exp(-t) - np.exp(-4*t)) * u(t)
```

```python
# Plot x(t), h(t), and h(-t)

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)

T = np.arange(-20.0, 20.01, 0.01)

axes[0].step(T, [x(t) for t in T])
axes[1].step(T, [h(t) for t in T])
axes[2].step(T, [h(-t) for t in T])

axes[0].set_title("Question: 1 Part (b)")
axes[0].set_ylabel("x(t)")
axes[1].set_ylabel("h(t)")
axes[2].set_ylabel("h(-t)")
axes[2].set_xlabel("t")
```

```python
# Plotting x(t), h(t), and y(t) = x(t) * h(t) for t = [0, 20]

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)

T = np.arange(0.0, 20.01, 0.01)
X = [x(t) for t in T]
H = [h(t) for t in T]

axes[0].plot(T, X)
axes[1].plot(T, H)
axes[2].plot(T, np.convolve(np.ravel(X), np.ravel(H), "same"))
# axes[2].plot(np.arange(0.0, 40.02, 0.01), np.convolve(np.ravel(X), np.ravel(H), "full"))

axes[0].set_title("Question: 1 Part (b)")
axes[0].set_ylabel("x(t)")
axes[1].set_ylabel("h(t)")
axes[2].set_ylabel("y(t)")
axes[2].set_xlabel("t")
```

```python
# Plotting x(t), h(t), and y(t) = x(t) * h(t) for t = [0, 20]

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)

T = np.linspace(0.0, 20.0, 1000)
X = [x(t) for t in T]
H = [h(t) for t in T]

axes[0].plot(T, X)
axes[1].plot(T, H)
axes[2].plot(T, np.convolve(np.ravel(X), np.ravel(H), "full")[:1000])

axes[0].set_title("Question: 1 Part (b)")
axes[0].set_ylabel("x(t)")
axes[1].set_ylabel("h(t)")
axes[2].set_ylabel("y(t)")
axes[2].set_xlabel("t")
```
