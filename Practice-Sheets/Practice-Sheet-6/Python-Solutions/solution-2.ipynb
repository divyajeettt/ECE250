```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = np.arange(-0.1, 0.10001, 0.0001)

# s(t): Sinusoid Signals

def s1(t: float) -> float:
    return 4 * np.sin(2*np.pi * 10 * t)

def s2(t: float) -> float:
    return 4 * np.sin(2*np.pi * 100 * t)

def s3(t: float) -> float:
    return s1(t) + s2(t)

def s4(t: float) -> float:
    return s1(t) - s2(t)

def s5(t: float) -> float:
    return s1(t) * s2(t)

def s6(t: float) -> float:
    return s3(t - 0.05)
```

```python
# (a) Frequency 10 Hz and Magnitude 4 units

plt.plot(T, [s1(t) for t in T])
plt.title("Sinusoid Signal with frequency 10 Hz and Magnitude 4 units")
plt.xlabel("t")
plt.ylabel("s1(t)")
plt.grid(True)
```

```python
# (b) Frequency 100 Hz and Magnitude 4 units

plt.plot(T, [s2(t) for t in T])
plt.title("Sinusoid Signal with frequency 100 Hz and Magnitude 4 units")
plt.xlabel("t")
plt.ylabel("s2(t)")
plt.grid(True)
```

```python
# (c) s3(t) = s1(t) + s2(t)

plt.plot(T, [s3(t) for t in T])
plt.title("s3(t) = s1(t) + s2(t)")
plt.xlabel("t")
plt.ylabel("s3(t)")
plt.grid(True)
```

```python
# (d) s4(t) = s1(t) - s2(t)

plt.plot(T, [s4(t) for t in T])
plt.title("s4(t) = s1(t) - s2(t)")
plt.xlabel("t")
plt.ylabel("s4(t)")
plt.grid(True)
```

```python
# (e) s5(t) = s1(t) s2(t)

plt.plot(T, [s5(t) for t in T])
plt.title("s5(t) = s1(t) s2(t)")
plt.xlabel("t")
plt.ylabel("s5(t)")
plt.grid(True)
```

```python
# (f) Time shifted Signal s6(t) = s3(t - 0.05)

fig, axes = plt.subplots(2, 1, figsize=(8, 7))
for i in range(2):
    axes[i].grid(True)

axes[0].plot(T, [s3(t) for t in T])
axes[1].plot(T, [s6(t) for t in T])

axes[0].set_title("s6(t) = s3(t - 0.05)")
axes[0].set_ylabel("s3(t)")
axes[1].set_ylabel("s6(t)")
axes[1].set_xlabel("t")
```
