```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sinc(t): Normalized Sinc Signal
def sinc(t: float) -> float:
    return np.sin(np.pi*t) / (np.pi*t)
```

```python
T = np.arange(-10.0, 10.0, 0.01)

plt.plot(T, [sinc(t) for t in T])

plt.title("Normalized Sinc Signal")
plt.ylabel("sinc(t)")
plt.xlabel("t")
plt.grid(True)
```
