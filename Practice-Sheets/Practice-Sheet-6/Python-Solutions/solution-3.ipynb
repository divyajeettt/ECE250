```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = np.arange(-5.0, 5.01, 0.01)

# r(t): Ramp Signal
def r(t: float) -> float:
    return t if t >= 0 else 0
```

```python
# (a) Time advancement and delay on r(t)

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)
    axes[i].set_ylim(-0.5, 4.5)

axes[0].plot(T, [r(t) for t in T])
axes[1].plot(T, [r(t+2.5) for t in T])
axes[2].plot(T, [r(t-2.5) for t in T])

axes[0].set_title("Performing Time Shifting (advancement and delay) on Ramp Signal")
axes[0].set_ylabel("r(t)")
axes[1].set_ylabel("r(t+a)")
axes[2].set_ylabel("r(t-a)")
axes[2].set_xlabel("t")
```

```python
# (b) Time reversal on r(t)

fig, axes = plt.subplots(2, 1, figsize=(8, 7))
for i in range(2):
    axes[i].grid(True)
    axes[i].set_ylim(-0.5, 4.5)

axes[0].plot(T, [r(t) for t in T])
axes[1].plot(T, [r(-t) for t in T])

axes[0].set_title("Performing Time Reversal on Ramp Signal")
axes[0].set_ylabel("r(t)")
axes[1].set_ylabel("r(-t)")
axes[1].set_xlabel("t")
```
