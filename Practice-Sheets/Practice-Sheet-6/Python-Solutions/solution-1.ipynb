```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A: float = 2.0
t1: float = -2.0
t2: float = 2.0

T = np.arange(-6, 6.01, 0.01)

# p(t): Rectangular Pulse Signal
def p(t: float) -> float:
    return A if t1 <= t <= t2 else 0
```

```python
# (a) Amplitude of 2 units from t = -2 to t = 2

p1 = [p(t) for t in T]
plt.step(T, p1)

plt.title("Rectangular Pulse Signal")
plt.xlabel("t")
plt.ylabel("p(t)")
plt.grid(True)
```

```python
# (b) The same signal with time scaling by a factor of 2

p2 = [p(2*t) for t in T]
plt.step(T, p2)

plt.title("Rectangular Pulse Signal")
plt.xlabel("t")
plt.ylabel("p(2t)")
plt.grid(True)
```

```python
# (c) The same signal with time scaling by a factor of 0.5

p3 = [p(t/2) for t in T]
plt.step(T, p3)

plt.title("Rectangular Pulse Signal")
plt.xlabel("t")
plt.ylabel("p(t/2)")
plt.grid(True)
```

```python
# (d) Generating p4(t) using the signals p1(t), p2(t), and p3(t)

p4 = [i+j+k for i, j, k in zip(p1, p2, p3)]
plt.step(T, p4)

plt.title("Signal p4(t)")
plt.xlabel("t")
plt.ylabel("p4(t)")
plt.grid(True)
```

```python
# (e) Convoling the signals p1(t) and p4(t)

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)

axes[0].plot(T, p1)
axes[1].plot(T, p4)
axes[2].plot(np.arange(-12.0, 12.01, 0.01), np.convolve(np.ravel(p1), np.ravel(p4), "full"))

axes[0].set_title("Convolution of p1(t) and p4(t)")
axes[0].set_ylabel("p1(t)")
axes[1].set_ylabel("p2(t)")
axes[2].set_ylabel("p1(t) * p2(t)")
axes[2].set_xlabel("t")
```
