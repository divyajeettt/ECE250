```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# s1(t): Sinusoid Signal 1
def s1(t: float) -> float:
    return 5 * np.sin(2*np.pi*1*t)

# s2(t): Sinusoid Signal 2
def s2(t: float) -> float:
    return 10 * np.sin(2*np.pi*2*t)
```

```python
T = np.arange(-1.0, 1.001, 0.001)

fig, axes = plt.subplots(3, 1, figsize=(8, 7))
for i in range(3):
    axes[i].grid(True)
    
S1 = [s1(t) for t in T]
S2 = [s2(t) for t in T]

axes[0].plot(T, S1)
axes[1].plot(T, S2)
axes[2].plot(np.arange(-2.0, 2.0001, 0.001), np.convolve(np.ravel(S1), np.ravel(S2), "full"))

axes[0].set_title("Convolution of Sinusoidal Signals")
axes[0].set_ylabel("sin(2t)")
axes[1].set_ylabel("sin(4t)")
axes[2].set_ylabel("sin(2t) * sin(4t)")
axes[2].set_xlabel("t")
```
