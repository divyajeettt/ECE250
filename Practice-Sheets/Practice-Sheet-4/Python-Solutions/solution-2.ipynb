```python
import matplotlib.pyplot as plt
import math
```

```python
# (a) Discrete-time cosine Signal

N = range(-25, 26)
plt.stem(N, [math.cos(n) for n in N], use_line_collection=True)

plt.title("Discrete-time cosine signal")
plt.xlabel("n")
plt.ylabel("cos[n]")
plt.grid(True)
```

```python
# (b) Discrete-time cosine Signal with time shift and scaling

N = range(-25, 26)
plt.stem(N, [math.cos(0.25*n - 5) for n in N], use_line_collection=True)

plt.title("Discrete-time cosine signal")
plt.xlabel("n")
plt.ylabel("cos[0.25n-5]")
plt.grid(True)
```
