```python
import matplotlib.pyplot as plt
import math
```

```python
# u[n]: Unit Step Signal
def u(n: int) -> int:
    return 1 if n >= 0 else 0

# x[n]: Signal given for Question 5
def x(n: int) -> float:
    return math.exp(-n/4) * u(n)
```

```python
# Plot x[n]

N = range(-10, 26)
X = [x(n) for n in N]
plt.stem(N, X, use_line_collection=True)

plt.title("Question: 5")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
```

```python
# Plot Odd{x[t]} and Even{x[n]}

fig, axes = plt.subplots(2, 1)
for i in range(2):
    axes[i].grid(True)

N = range(-20, 21)
oddX = [(x(n) - x(-n))/2 for n in N]
evenX = [(x(n) + x(-n))/2 for n in N]

axes[0].stem(N, oddX, use_line_collection=True)
axes[1].stem(N, evenX, use_line_collection=True)

axes[0].set_title("Odd and Even parts of x[n]")
axes[0].set_ylabel("Odd{x[n]}")
axes[1].set_ylabel("Even{x[n]}")
axes[1].set_xlabel("n")
```
