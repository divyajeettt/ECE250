```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
theta: float = cmath.pi / 2
C: int = 2
a: float = 0.2
b: float = 3.0

# x(t): Signal given for Question 6
def x(t: float) -> complex:
    return C * cmath.exp(complex(0, theta)) * cmath.exp(complex(a, b)*t)
```

```python
# Plot Real{x(t)} and Imag{x(t)}

fig, axes = plt.subplots(2, 1)
for i in range(2):
    axes[i].grid(True)

T = np.arange(-1, 26.6, 0.1)
realX = [x(t).real for t in T]
imagX = [x(t).imag for t in T]

axes[0].plot(T, realX)
axes[1].plot(T, imagX)

axes[0].set_title("Real and Imaginary parts of x[n]")
axes[0].set_ylabel("Real{x(t)}")
axes[1].set_ylabel("Imag{x(t)}")
axes[1].set_xlabel("t")
```
