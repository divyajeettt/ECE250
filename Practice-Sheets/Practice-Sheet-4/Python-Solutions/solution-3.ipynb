```python
import matplotlib.pyplot as plt
import math
```

```python
# (a) The signal will be periodic

N = range(-50, 51)
plt.stem(N, [math.cos(2*math.pi*n/5) * math.cos(2*math.pi*n/7) for n in N], use_line_collection=True)

plt.title("Question: 3 (a)")
plt.xlabel("n")
plt.ylabel("cos[2πn/5] cos[2πn/7]")
plt.grid(True)
```

```python
# (b) The signal will be aperiodic

N = range(-50, 51)
plt.stem(N, [math.sin(math.pi + 0.2*n) for n in N], use_line_collection=True)

plt.title("Question: 3 (b)")
plt.xlabel("n")
plt.ylabel("sin[π + 0.2n]")
plt.grid(True)
```
