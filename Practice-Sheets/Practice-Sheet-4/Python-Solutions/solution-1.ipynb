```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# u[n]: Unit Step Signal
def u(n: int) -> int:
    return 1 if n >= 0 else 0
```

```python
# (a) Continuous-time Unit Step Signal

X = np.arange(-10, 10.01, 0.01)

plt.step(X, [u(t) for t in X], linewidth=2)

plt.title("Continuous-time Unit Step Signal")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.grid(True)
```

```python
# (b) Discrete-time Unit Step Signal

N = range(-10, 11)

plt.stem(N, [u(n) for n in N], use_line_collection=True)

plt.title("Discrete-time Unit Step Signal")
plt.xlabel("n")
plt.ylabel("u[n]")
plt.grid(True)
```

```python
# (c) Continuous-time Unit Step Signal with time shift

X = np.arange(-10, 10.01, 0.01)

plt.step(X, [u(t-2) for t in X], linewidth=2)

plt.title("Continuous-time Unit Step Signal with time shift")
plt.xlabel("t")
plt.ylabel("u(t-2)")
plt.grid(True)
```

```python
# (d) Discrete-time Unit Step Signal with time shift

N = range(-10, 11)

plt.stem(N, [u(n-2) for n in N], use_line_collection=True)

plt.title("Discrete-time Unit Step Signal with time shift")
plt.xlabel("n")
plt.ylabel("u[n-2]")
plt.grid(True)
```
