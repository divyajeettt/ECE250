```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
theta: float = cmath.pi / 2
C: int = 2
a: float = 0.2
b: float = 3.0

# x(t): Signal given for Question 6
def x(t: float) -> complex:
    return C * cmath.exp(complex(0, theta)) * cmath.exp(complex(a, b)*t)
```

```python
# Plot Real{x(t)}

T = np.arange(-1, 26.6, 0.1)
realX = [x(t).real for t in T]

plt.plot(T, realX)
plt.grid(True)

plt.title("Real part of x(t)")
plt.ylabel("Real{x(t)}")
plt.xlabel("t")
```

```python
# Plot Imag{x(t)}

T = np.arange(-1, 26.6, 0.1)
imagX = [x(t).imag for t in T]

plt.plot(T, imagX)
plt.grid(True)

plt.title("Imaginary part of x(t)")
plt.ylabel("Imag{x(t)}")
plt.xlabel("t")
```
