```python
import matplotlib.pyplot as plt
import math
```

```python
# u[n]: Unit Step Signal
def u(n: int) -> int:
    return 1 if n >= 0 else 0

# x[n]: Signal given for Question 5
def x(n: int) -> float:
    return math.exp(-n/4) * u(n)
```

```python
# Plot x[n]

N = range(-10, 26)
X = [x(n) for n in N]
plt.stem(N, X, use_line_collection=True)

plt.title("Question: 5")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
```

```python
# Plot Odd{x[n]}

N = range(-20, 21)
oddX = [(x(n) - x(-n))/2 for n in N]

plt.stem(N, oddX, use_line_collection=True)
plt.grid(True)

plt.title("Odd part of x[n]")
plt.ylabel("Odd{x[n]}")
plt.xlabel("n")
```

```python
# Plot Even{x[n]}

N = range(-20, 21)
evenX = [(x(n) + x(-n))/2 for n in N]

plt.stem(N, evenX, use_line_collection=True)
plt.grid(True)

plt.title("Even part of x[n]")
plt.ylabel("Even{x[n]}")
plt.xlabel("n")
```
