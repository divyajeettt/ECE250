```python
import matplotlib.pyplot as plt
import math
```

```python
# u[n]: Unit Step Signal
def u(n: int) -> int:
    return 1 if n >= 0 else 0

# x[n]: Signal for Q5, part (i)
def x(n: int) -> float:
    return n * (u(n)-u(n-10)) + 10*math.exp(-0.3*(n-10))*(u(n-10)-u(n-20))

# y[n]: Signal for Q5, part (ii)
def y(n: int) -> float:
    return math.cos(0.03 * math.pi * n) + u(n)

# z[n]: Signal for Q6
def z(n: int) -> int:
    return u(n) - u(n-10)
```

```python
# Plot x[n]

N = range(0, 21)
X = [x(n) for n in N]
plt.stem(N, X)

plt.title("Question-5 Part (i)")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
```

```python
# Plot y[n]

N = range(0, 51)
Y = [y(n) for n in N]
plt.stem(N, Y)

plt.title("Question-5 Part (ii)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
```

```python
# Plot z[n], Odd{z[n]}, and Even{z[n]}

N = range(-20, 21)
Z = [z(n) for n in N]
oddZ = [(z(n) - z(-n)) / 2 for n in N]
evenZ = [(z(n) + z(-n)) / 2 for n in N]

fig, axes = plt.subplots(3, 1)

for i in range(3):
    axes[i].grid(True)

axes[0].stem(N, Z)
axes[1].stem(N, oddZ)
axes[2].stem(N, evenZ)

axes[0].set_title("Question-6")
axes[0].set_ylabel("z[n]")
axes[1].set_ylabel("Odd{z[n]}")
axes[2].set_ylabel("Even{z[n]}")
axes[2].set_xlabel("n")
```
